Part A: Graphs using MatPlot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = {
    'Days': ['Mon','Tue','Wed','Thu','Fri'],
    'Study_Hours': [2,3,2.5,4,3.5],
    'Marks':[50,60,55,70,65],
    'Attendance': [60,70,65,80,75],
    'Sleep_Hours': [6,7,6,8,7],
    'Assignments_Completed': [1,2,1,3,2]
}

df = pd.DataFrame(data)
display(df)
```

```python
#1.1 Line Chart (Trend)

plt.plot(df['Days'],df['Study_Hours'],marker='o')
plt.title("Nakshatra")
plt.xlabel("Days")
plt.ylabel("Study Hours")
plt.show()
```

```python
#1.2. Advanced Line Chart

plt.figure(figsize=(7,4))

plt.plot(df['Days'],df['Study_Hours'],
         marker='o',color='green',label='Study Hours')

plt.plot(df['Days'],df['Marks'],
         marker='s',color='red',label='Marks')

plt.title("Nakshatra")
plt.xlabel("Days")
plt.ylabel("Values")
plt.legend()
plt.show()

```

```python
#1.3. Bar Graph

from matplotlib.lines import lineStyles
plt.bar(df['Days'],df['Marks'],color='green',edgecolor='black')

plt.title("Nakshatra")
plt.xlabel("Marks")
plt.ylabel("Values")
plt.grid(axis='x',linestyle='--', alpha = 0.5)
plt.grid(axis='y',linestyle='--', alpha = 0.5)
plt.show()
```

```python
#1.4. Advanced Bar Graph

bars = plt.bar(df['Days'], df['Marks'],color = 'green',edgecolor='black')
for bar in bars:
  y = bar.get_height()
  plt.text(bar.get_x()+bar.get_width()/2,y,str(y),ha='center',va='bottom')

plt.title("Nakshatra")
plt.xlabel("Marks")
plt.ylabel("Values")
plt.grid(axis='x',linestyle='--', alpha = 0.5)
plt.grid(axis='y',linestyle='--', alpha = 0.5)
plt.show()
```

```python
#1.5. Advanced Bar Graph (Multi)

import numpy as np
import matplotlib.pyplot as plt
x = np.arange(len(df['Days']))
width = 0.40
plt.bar( x - width/2, df['Study_Hours'], width, label = 'Study Hours')
plt.bar( x + width/2, df['Marks'], width, label = 'Marks')
plt.xticks(x, df['Days'])
plt.title("Nakshatra")
plt.xlabel("Days")
plt.ylabel("Marks")
plt.grid(axis='x',linestyle='--', alpha = 0.5)
plt.grid(axis='y',linestyle='--', alpha = 0.5)
plt.legend()
plt.show()
```

```python
#1.6. Histogram

import matplotlib.pyplot as plt
import numpy as np

marks = [50,60,55,70,65,75,80,85,90,45,68,72,77,82,88]

mean_value = np.mean(marks)
plt.figure(figsize=(7,4))
plt.hist(marks, bins=3, color='skyblue', edgecolor='black')
plt.axvline(mean_value, color='red', linestyle= '--', linewidth=2, label='Mean')
plt.title("Nakshatra")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.grid(axis='x',linestyle='--', alpha = 0.5)
plt.grid(axis='y',linestyle='--', alpha = 0.5)
plt.legend()
plt.show()
```

```python
#1.7. Scatter Plot (Relationship)

import matplotlib.pyplot as plt
plt.scatter(df['Study_Hours'],df['Marks'])
plt.title("Nakshatra")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.grid(axis='x',linestyle='--', alpha = 0.5)
plt.grid(axis='y',linestyle='--', alpha = 0.5)
plt.show()
```

```python
#1.8. Conditional Scatter Plot

import matplotlib.pyplot as plt
df['Result']=['Fail','Pass','Fail','Pass','Pass']
colors = ['red' if r == 'Fail' else 'green' for r in df['Result']]
plt.scatter(df['Study_Hours'],df['Marks'], c=colors)
plt.title("Nakshatra")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.grid(axis='x',linestyle='--', alpha = 0.5)
plt.grid(axis='y',linestyle='--', alpha = 0.5)
plt.show()
```

Part B: Graphs using Seaborn

```python
data1 = {
    'Day': ['Mon','Tue','Wed','Thu','Fri','Sat','Sun','Mon','Tue','Wed'],
    'Region': ['North','South','East','West','North','South','East','West','North',
               'South'],
    'Sales': [200,220,250,270,300,350,400,280,260,290],
    'Profit': [20,25,30,35,50,60,80,40,38,45],
    'Customers': [30,35,40,45,50,65,80,55,48,52],
    'Category': ['Food','Clothing','Electronics','Food','Clothing','Electronics',
                 'Food','Clothing','Electronics','Food']
}
df1 = pd.DataFrame(data1)
display(df1)
```

```python
#2.1 Line Plot

sns.lineplot(x= 'Day', y= 'Sales', data = df1, errorbar=None)
plt.title('Nakshatra')
plt.show()
```

```python
#2.2 Bar Plot

sns.barplot(x= 'Day', y= 'Sales', data = df1, errorbar=None)
plt.title('Nakshatra')
plt.show()
```

```python
#2.3 Histogram

sns.histplot(df1['Sales'], bins=5)
plt.title('Nakshatra')
plt.show()
```

```python
#2.4 Scatter Plot
sns.scatterplot(x='Sales',y='Profit',data=df1)
plt.title("Nakshatra")
plt.show()
```
